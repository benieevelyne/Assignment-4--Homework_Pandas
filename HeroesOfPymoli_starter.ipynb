```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# Raw data file
file_to_load = "purchase_data.csv"

# Read purchasing file and store into pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head(10)
```

## Player Count

* Display the total number of players


```python
Total_Number_of_Players = len(purchase_data["SN"].unique())
Total_Number_of_Players
Total = pd.DataFrame({"Total of Players": Total_Number_of_Players}, index=[0])
Total
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Number of Unique items
Number_of_Unique_Items = len(purchase_data["Item ID"].unique())

# Average Price
AveragePrice_Not_Rounded = purchase_data["Price"].mean()
Average_Price_Rounded = round(AveragePrice_Not_Rounded, 2)

#Number of Purchases
Purchases_count = purchase_data["Purchase ID"].count()

#Total Revenue
Total_Revenue = AveragePrice_Not_Rounded*Purchases_count

#Creating a data frame
df = pd.DataFrame({'Number of Unique Items': [Number_of_Unique_Items], 'Average Price': [str('$'+str(Average_Price_Rounded))], 
                   'Number of Purchases': [Purchases_count], 'Total Revenues': [Total_Revenue]})


df["Total Revenues"] = df["Total Revenues"].map("${:,.2f}".format)
df
```

## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python

#Purchase count per gender
male_purchase_count = purchase_data['Gender'].value_counts()['Male']
female_purchase_count = purchase_data['Gender'].value_counts()['Female']
other_purchase_count = Purchases_count - male_purchase_count - female_purchase_count

#Average Purchase Price and Total Purchase
male_purchase = purchase_data.loc[purchase_data['Gender'] == 'Male', :]
male_avg_prices = male_purchase['Price'].mean()
male_avg_prices_Rounded = round(male_avg_prices, 2)

male_total_purchase = male_purchase.sum()["Price"]


female_purchase = purchase_data.loc[purchase_data['Gender'] == 'Female', :]
female_avg_prices = female_purchase['Price'].mean()
female_avg_prices_Rounded = round(female_avg_prices, 2)

female_total_purchase = female_purchase.sum()["Price"]


other_purchase_less_female = purchase_data.loc[purchase_data['Gender'] != 'Female', :]
other_purchase = other_purchase_less_female.loc[other_purchase_less_female['Gender'] != 'Male', :]
other_avg_price = other_purchase['Price'].mean()

other_avg_prices = other_purchase['Price'].mean()
other_avg_prices_Rounded = round(other_avg_prices, 2)

other_total_purchase = other_purchase.sum()["Price"]

# Avg Purchase Total per Person
male_Avg_Purchase_Total_per_Person = round(male_total_purchase/male_purchase_count, 2)
female_Avg_Purchase_Total_per_Person = round(female_total_purchase/female_purchase_count, 2)
other_Avg_Purchase_Total_per_Person = round(other_total_purchase/other_purchase_count, 2)
    
#Summarize results in a dataframe
df_calcs2 = {'Purchase Count':[female_purchase_count,male_purchase_count,other_purchase_count],'Gender':['Female','Male','Other/Non-Disclosed'],'Average Purchase Price':[female_avg_prices_Rounded,male_avg_prices_Rounded,other_avg_prices_Rounded], 'Total Purchase Value':[female_total_purchase,male_total_purchase,other_total_purchase], 'Avg Purchase Total per Person':[female_Avg_Purchase_Total_per_Person,male_Avg_Purchase_Total_per_Person,other_Avg_Purchase_Total_per_Person] }
df_new2 =  pd.DataFrame(df_calcs2)
df_new2 = df_new2.set_index('Gender')

df_new2['Average Purchase Price'] = df_new2['Average Purchase Price'].map("${:,.2f}".format)
df_new2['Total Purchase Value'] = df_new2['Total Purchase Value'].map("${:,.2f}".format)
df_new2['Avg Purchase Total per Person'] = df_new2['Avg Purchase Total per Person'].map("${:,.2f}".format)
df_new2 = df_new2[['Purchase Count','Average Purchase Price', 'Total Purchase Value', 'Avg Purchase Total per Person']]

df_new2
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Establish bins for ages
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

#Categorize the existing players using the age bins. Hint: use pd.cut()
player_demographics = purchase_data.loc[:, ["Gender", "SN", "Age"]]

player_demographics["Age ranges"] = pd.cut(player_demographics["Age"], age_bins, labels=group_names)
player_demographics.head()
```

```python
#Calculate the numbers and percentages by age group
age_demographics_total = player_demographics["Age ranges"].value_counts()
age_demographics_percentage = round(age_demographics_total / Total_Number_of_Players * 100, 2)


#Create a summary data frame to hold the results
age_demographics_df = pd.DataFrame({"Percentage of Players": age_demographics_percentage, "Total Count": age_demographics_total})

#Optional: round the percentage column to two decimal points
age_demographics_df['Percentage of Players'] = age_demographics_df['Percentage of Players']

age_demographics_df = age_demographics_df.sort_index()
#Display Age Demographics Table
age_demographics_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Purchasing Analysis (Age)
player_demographics_with_prices = purchase_data.drop_duplicates(subset='SN', keep="first")
player_demographics_with_prices = purchase_data.loc[:, ["SN", "Price", "Age"]]
# Setting the index values
player_demographics_with_prices['Age ranges'] = pd.cut(purchase_data['Age'], age_bins, labels = group_names)



pur_age_group = player_demographics_with_prices.groupby(['Age ranges'])
pur_age_count = pur_age_group.size()
pur_age_avg_price = round(pur_age_group["Price"].mean(),2)
pur_age_tot_value = pur_age_group["Price"].sum()
pur_age_norm_tot = round(pur_age_tot_value/pur_age_count,2)
pur_age_dataframe = pd.DataFrame({"Purchase Count":pur_age_count,
                             "Average Purchase Price": pur_age_avg_price,
                             "Total Purchase Value": pur_age_tot_value,
                             "Average Purchase Total per Person": pur_age_norm_tot})

pur_age_dataframe['Average Purchase Price'] = pur_age_dataframe['Average Purchase Price'].map("${:.2f}".format)
pur_age_dataframe['Total Purchase Value'] = pur_age_dataframe['Total Purchase Value'].map("${:.2f}".format)
pur_age_dataframe['Average Purchase Total per Person'] = pur_age_dataframe['Average Purchase Total per Person'].map("${:.2f}".format)


#age_dataframe = age_dataframe.set_index('Age ranges')

#age_dataframe = age_dataframe.sort_values('Total Purchase Value', ascending = False)
age_dataframe
```
